```python
# IMDb Yorumları Sınıflandırmak İçin GAN + LSTM Modeli

# Gerekli kütüphanelerin yüklenmesi
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding, Flatten, Reshape
import numpy as np
import matplotlib.pyplot as plt
import random

# 1. IMDb veri setinin yüklenmesi
num_words = 10000  # En sık kullanılan 10,000 kelimeyi tutuyoruz
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=num_words)

# IMDb kelime indeksini çözmek için sözlük
word_index = imdb.get_word_index()
reverse_word_index = {value: key for (key, value) in word_index.items()}

def decode_review(encoded_review):
    """Encoded yorumu okunabilir kelime formatına çevirir."""
    return ' '.join([reverse_word_index.get(i - 3, '?') for i in encoded_review])

# Eğitim setinden örnek 5 yorumu ve etiketleri yazdır
print("Eğitim veri setinden 5 örnek yorum:")
for i in range(5):
    print(f"\nÖrnek {i + 1}:")
    print("Yorum:", decode_review(X_train[i]))
    print("Etiket:", "Olumlu" if y_train[i] == 1 else "Olumsuz")

# 2. Veri ön işleme
max_len = 200  # Her yorumu 200 kelimeyle sınırlıyoruz
X_train = pad_sequences(X_train, maxlen=max_len, padding='post')
X_test = pad_sequences(X_test, maxlen=max_len, padding='post')

# 3. GAN Modelinin Kurulması
embedding_dim = 128
latent_dim = 100  # Rastgele gürültü boyutu

# Generator: Gürültüden sahte IMDb yorumları üretir
def build_generator():
    model = Sequential([
        Dense(256, activation='relu', input_dim=latent_dim),  # Rastgele gürültüyü işliyor
        Dense(max_len, activation='sigmoid'),  # Yorum uzunluğunu elde et
        Reshape((max_len, 1))  # LSTM için uygun şekil
    ])
    return model

generator = build_generator()
print("GAN Generator Summary:")
generator.summary()

# Discriminator: Gelen verinin gerçek mi sahte mi olduğunu ayırt eder
def build_discriminator():
    model = Sequential([
        Flatten(input_shape=(max_len, 1)),  # Veriyi düzleştirir
        Dense(128, activation='relu'),
        Dense(1, activation='sigmoid')  # Çıkış: gerçek (1) veya sahte (0)
    ])
    return model

discriminator = build_discriminator()
discriminator.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print("GAN Discriminator Summary:")
discriminator.summary()

# GAN Modeli: Generator ve Discriminator'ı birleştirir
def build_gan(generator, discriminator):
    discriminator.trainable = False  # Discriminator'u GAN eğitimi sırasında donduruyoruz
    model = Sequential([generator, discriminator])
    model.compile(optimizer='adam', loss='binary_crossentropy')
    return model

gan = build_gan(generator, discriminator)
gan.summary()

# 4. GAN Modelinin Eğitilmesi
epochs = 5
batch_size = 32
real_labels = np.ones((batch_size, 1))
fake_labels = np.zeros((batch_size, 1))

def train_gan(epochs, batch_size):
    for epoch in range(epochs):
        # 1. Gerçek veri seçimi
        idx = np.random.randint(0, X_train.shape[0], batch_size)
        real_reviews = X_train[idx]
        real_reviews = real_reviews.reshape((batch_size, max_len, 1))

        # 2. Sahte veri üretimi
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        fake_reviews = generator.predict(noise, verbose=0)

        # 3. Discriminator'ı eğit
        d_loss_real = discriminator.train_on_batch(real_reviews, real_labels)[0]  # Sadece loss değeri al
        d_loss_fake = discriminator.train_on_batch(fake_reviews, fake_labels)[0]  # Sadece loss değeri al
        d_loss = 0.5 * (d_loss_real + d_loss_fake)

        # 4. Generator'ı eğit
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        g_loss = gan.train_on_batch(noise, real_labels)[0]  # Sadece loss değeri al

        # Her epoch sonunda kayıpları yazdır
        print(f"Epoch {epoch + 1}/{epochs}, D Loss: {d_loss:.4f}, G Loss: {g_loss:.4f}")

print("GAN Training:")
train_gan(epochs, batch_size)

# 5. LSTM Sınıflandırıcı Modeli
model = Sequential([
    Embedding(input_dim=num_words, output_dim=128, input_length=max_len),
    LSTM(128, return_sequences=False),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print("LSTM Summary:")
model.summary();

history = model.fit(X_train, y_train, epochs=5, batch_size=32, validation_split=0.2)

# 6. Eğitim ve doğrulama sonuçlarının grafiği
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Eğitim Kaybı')
plt.plot(history.history['val_loss'], label='Doğrulama Kaybı')
plt.title('Eğitim ve Doğrulama Kaybı')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Eğitim Doğruluğu')
plt.plot(history.history['val_accuracy'], label='Doğrulama Doğruluğu')
plt.title('Eğitim ve Doğrulama Doğruluğu')
plt.legend()

plt.show()

# 7. Rastgele Test Verisinden 5 Örnek Tahmin
print("\nTest veri setinden rastgele 5 örnek yorum ve tahminleri:")
for i in range(5):
    idx = random.randint(0, len(X_test) - 1)
    test_review = X_test[idx]
    true_label = "Olumlu" if y_test[idx] == 1 else "Olumsuz"
    prediction = model.predict(np.expand_dims(test_review, axis=0))[0][0]
    predicted_label = "Olumlu" if prediction >= 0.5 else "Olumsuz"
    print(f"\nÖrnek {i + 1}:")
    print("Gerçek Etiket:", true_label)
    print("Tahmin Edilen Etiket:", predicted_label, f"(Olasılık: {prediction:.4f})")

    # 8. GAN Tarafından Üretilen 5 Yorumun LSTM ile Test Edilmesi
    # GAN tarafından sahte yorum üretilmesi ve LSTM tahmini
print("\nGAN tarafından üretilen yorumların LSTM modeli ile sınıflandırılması:")
noise = np.random.normal(0, 1, (5, latent_dim))  # 5 örnek için rastgele gürültü oluştur
fake_reviews = generator.predict(noise)
for i, review in enumerate(fake_reviews):
    # Üretilen yorumu test etmek için LSTM modeline uygun hale getir
    reshaped_review = review.reshape(1, max_len)
    prediction = model.predict(reshaped_review, verbose=0)[0][0]
    predicted_label = "Olumlu" if prediction >= 0.5 else "Olumsuz"
    decoded_review = decode_review([int(round(word[0] * num_words)) for word in review[:max_len]])
    print(f"\nÜretilen Yorum {i + 1}:")
    print(decoded_review)
    print("LSTM Tahmini:", predicted_label, f"(Olasılık: {prediction:.4f})")
```
